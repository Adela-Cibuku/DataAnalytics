```python
from fuzzywuzzy import process

# Listing of standardized city names 
standardized_cities = ["Santa Barbara", "Affton", "Tucson", "Philadelphia", "Green Lane", "Ashland City", "Brentwood", "St. Petersburg", "Nashville"]

# Function to find the best match for each city
def normalize_city(city, choices, threshold=80):
    match = process.extractOne(city, choices)
    if match and match[1] >= threshold:
        return match[0]  # Return the matched city
    return city  # If no match is found or the match is below threshold, return the original

# Normalizing city names using fuzzy matching
df_pandas['normalized_city'] = df_pandas['city'].apply(lambda x: normalize_city(x, standardized_cities))

# Displaying the result
print(df_pandas[['city', 'normalized_city']].drop_duplicates())
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, regexp_replace

# Initialize Spark Session
spark = SparkSession.builder.appName("YelpDataCleaning").getOrCreate()

# Load the dataset
df = spark.read.json("yelp_academic_dataset_business.json")

# Function to standardize time format
def standardize_time(column_name):
    return (
        regexp_replace(column_name, r"(\d{1,2})\s*am", r"$1:00")  # Convert am times
    )

# Standardize hours for each day
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

for day in days:
    # Access the nested field with getField() and apply the transformation
    df = df.withColumn(f"{day}", standardize_time(col("hours").getField(day)))

# Display to check the changes
df.select("Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday").show(5, truncate=False)
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import (
    col, when, split, concat_ws, lit, lpad, trim, regexp_extract
)

# Initialize Spark Session
spark = SparkSession.builder.appName("YelpDataCleaning").getOrCreate()

# (Assuming df is already loaded as before)

# Function to standardize a single time value (e.g., '8:0' â†’ '08:00')
def standardize_time(time_str_col):
    # Ensure the time string is trimmed
    t_str = trim(time_str_col)
    # Extract hours and minutes using regex with end-of-string anchor for minutes
    extracted_hour = regexp_extract(t_str, r"^(\d+):", 1)
    extracted_minute = regexp_extract(t_str, r":(\d+)$", 1)
    
    # If extraction fails, we return None
    standardized = when(
        (extracted_hour == "") | (extracted_minute == ""), 
        None
    ).otherwise(
        concat_ws(":", lpad(extracted_hour, 2, "0"), lpad(extracted_minute, 2, "0"))
    )
    return standardized

# Function to standardize the time range for a day
def standardize_time_range(time_range_col):
    split_times = split(time_range_col, "-")
    start_time = standardize_time(split_times.getItem(0))
    end_time = standardize_time(split_times.getItem(1))
    time_range_std = concat_ws("-", start_time, end_time)
    
    return when(
        time_range_col.isNull(), 
        lit("Closed")
    ).when(
        start_time.isNull() | end_time.isNull(), 
        lit("Closed")
    ).otherwise(
        when(time_range_std == "00:00-00:00", lit("Closed")).otherwise(time_range_std)
    )


# List of days
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

# IMPORTANT: Use the original nested "hours" struct for transformation.
# This avoids re-processing columns that were already modified.
for day in days:
    df = df.withColumn(day, standardize_time_range(col("hours").getField(day)))

# Show the updated day columns
df.select(days).show(5, truncate=False)
```

```python
from pyspark.sql.functions import col, sum, when, lit
from pyspark.sql.types import StringType

# Build a list of expressions to count missing values per column.
cols_missing = []
for field in df.schema.fields:
    col_name = field.name
    # If the column is a string, check both for null and empty string.
    if isinstance(field.dataType, StringType):
        condition = (col(col_name).isNull() | (col(col_name) == ""))
    else:
        # For non-string columns, just check for null.
        condition = col(col_name).isNull()
    
    cols_missing.append(sum(when(condition, 1).otherwise(0)).alias(col_name))

# Create a DataFrame that shows the count of missing values per column.
missing_values_df = df.select(*cols_missing)
missing_values_df.show(truncate=False)
```

```python
from pyspark.sql import functions as F
from pyspark.sql.types import StringType, StructType

# Set your missing-value threshold (e.g., 50%)
threshold = 0.5

# Count the total number of rows in the DataFrame
total_count = df.count()

# List to store expressions for missing value count
exprs = []
# Also keep track of which columns we are evaluating (only atomic columns)
cols_to_consider = []

for field in df.schema.fields:
    col_name = field.name
    # Only consider atomic (non-struct) columns.
    if not isinstance(field.dataType, StructType):
        cols_to_consider.append(col_name)
        # For string columns, also check for empty strings.
        if isinstance(field.dataType, StringType):
            expr = F.sum(
                F.when(F.col(col_name).isNull() | (F.col(col_name) == ""), 1).otherwise(0)
            ).alias(col_name)
        else:
            expr = F.sum(
                F.when(F.col(col_name).isNull(), 1).otherwise(0)
            ).alias(col_name)
        exprs.append(expr)
    else:
        # For nested columns, you might choose to ignore them (or flatten them first)
        print(f"Skipping nested column: {col_name}")

# Create a DataFrame with the missing counts for the considered columns
missing_values_df = df.select(*exprs)

# Collect the missing counts as a dictionary
missing_counts = missing_values_df.collect()[0].asDict()

# Identify columns where the fraction of missing values exceeds the threshold
columns_to_drop = [
    col_name for col_name, missing_count in missing_counts.items() 
    if (missing_count / total_count) > threshold
]

print("Dropping columns:", columns_to_drop)

# Drop the identified columns from the DataFrame
df = df.drop(*columns_to_drop)

# Show the schema after dropping columns
df.printSchema()
```

```python
from pyspark.sql.functions import col, sum, when
from pyspark.sql.types import StringType, StructType

# Build expressions to count missing values for each atomic column
missing_exprs = []
for field in df.schema.fields:
    col_name = field.name
    # Skip nested columns (like the original 'hours' struct)
    if isinstance(field.dataType, StructType):
        print(f"Skipping nested column: {col_name}")
        continue

    # For string columns, count nulls and empty strings; for others, just nulls.
    if isinstance(field.dataType, StringType):
        expr = sum(when(col(col_name).isNull() | (col(col_name) == ""), 1).otherwise(0)).alias(col_name)
    else:
        expr = sum(when(col(col_name).isNull(), 1).otherwise(0)).alias(col_name)
    
    missing_exprs.append(expr)

# Compute missing counts for each column
missing_counts_df = df.select(*missing_exprs)
missing_counts_df.show(truncate=False)
```

```python
from pyspark.sql.functions import col, when, sum

# Initialize a dictionary to store missing value counts for each column
missing_counts = {}

# Check each column for missing values
for c in df.columns:
    if isinstance(df.schema[c].dataType, StringType):
        missing_count = df.filter(col(c).isNull() | (col(c) == "")).count()
    else:
        missing_count = df.filter(col(c).isNull()).count()
    
    missing_counts[c] = missing_count

# Display missing value counts for each column
missing_counts
```

```python
from pyspark.sql.functions import when, lit, col, struct

# Fill missing values for struct columns (hours) by filling individual day parts
df_cleaned = df.withColumn(
    "hours",
    when(col("hours").isNull(),
         struct(
             lit("Unknown").alias("Monday"),
             lit("Unknown").alias("Tuesday"),
             lit("Unknown").alias("Wednesday"),
             lit("Unknown").alias("Thursday"),
             lit("Unknown").alias("Friday"),
             lit("Unknown").alias("Saturday"),
             lit("Unknown").alias("Sunday")
         )
    ).otherwise(
        struct(
            when(col("hours.Monday").isNull(), lit("Unknown")).otherwise(col("hours.Monday")).alias("Monday"),
            when(col("hours.Tuesday").isNull(), lit("Unknown")).otherwise(col("hours.Tuesday")).alias("Tuesday"),
            when(col("hours.Wednesday").isNull(), lit("Unknown")).otherwise(col("hours.Wednesday")).alias("Wednesday"),
            when(col("hours.Thursday").isNull(), lit("Unknown")).otherwise(col("hours.Thursday")).alias("Thursday"),
            when(col("hours.Friday").isNull(), lit("Unknown")).otherwise(col("hours.Friday")).alias("Friday"),
            when(col("hours.Saturday").isNull(), lit("Unknown")).otherwise(col("hours.Saturday")).alias("Saturday"),
            when(col("hours.Sunday").isNull(), lit("Unknown")).otherwise(col("hours.Sunday")).alias("Sunday")
        )
    )
)

# Fill missing values for other columns
df_cleaned = df_cleaned.fillna({
    'address': 'Unknown',
    'categories': 'Unknown',
    'postal_code': 'Unknown',
    'BusinessAcceptsCreditCards': 'Unknown',
    'BusinessParking': 'Unknown',
    'RestaurantsPriceRange2': 'Unknown',
    'review_count': 0
})

# For numeric columns like 'stars', you can fill with the mean (if needed)
df_cleaned = df_cleaned.fillna({
    'stars': df_cleaned.agg({'stars': 'mean'}).collect()[0][0]  # Fill 'stars' with the mean value
})

# Now, let's check for missing values by handling struct and non-struct columns separately
missing_counts = {}

# Handle struct columns (e.g., 'hours') separately
for c in df_cleaned.columns:
    if isinstance(df_cleaned.schema[c].dataType, StructType):  # Check if the column is a struct
        missing_count = df_cleaned.filter(col(c).isNull()).count()
    else:
        missing_count = df_cleaned.filter(col(c).isNull() | (col(c) == "")).count()
    
    missing_counts[c] = missing_count

# Display the missing value counts for each column
missing_counts
```

```python
from pyspark.sql.functions import coalesce, lit

# Fill missing values for address and postal_code with "Unknown" or any other placeholder
df_cleaned = df_cleaned.withColumn(
    'address', coalesce(col('address'), lit('Unknown'))
).withColumn(
    'postal_code', coalesce(col('postal_code'), lit('Unknown'))
)

# Check missing counts after replacement
missing_counts = {}
for c in df_cleaned.columns:
    missing_count = df_cleaned.filter(col(c).isNull()).count()
    missing_counts[c] = missing_count

# Display the missing value counts for each column
missing_counts

```

```python
# If not already defined, load the dataset and convert to Pandas
from pyspark.sql import SparkSession
import pandas as pd

# Initialize Spark session
spark = SparkSession.builder.appName("YelpDataAnalysis").getOrCreate()

# Load dataset
df = spark.read.json("DA_Project.json")

# Convert to Pandas DataFrame
df_pandas = df.toPandas()

# Enhance Correlation Analysis with More Attributes
num_cols = ["stars", "review_count", "is_open", "latitude", "longitude"]
corr_matrix = df_pandas[num_cols].corr()
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Enhanced Correlation Heatmap with Additional Attributes")
plt.show()
```

```python
# DATA ENRICHMENT

# 1. Feature Engineering - Popularity Score
df_pandas["popularity_score"] = df_pandas["review_count"] * df_pandas["stars"]

# 2. Business Status - Open/Closed
df_pandas["is_open_binary"] = df_pandas["is_open"].apply(lambda x: "Open" if x == 1 else "Closed")

# 3. City-wise Business Density
df_pandas["business_density"] = df_pandas.groupby("city")["city"].transform("count")

# Visualizing Popularity Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df_pandas["popularity_score"], bins=30, kde=True, color="purple")
plt.title("Distribution of Popularity Scores")
plt.xlabel("Popularity Score")
plt.show()
```
